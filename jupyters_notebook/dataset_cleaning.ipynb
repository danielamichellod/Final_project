```python
# Importations

import pandas as pd 
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import sys
sys.path.append("../SRC")

import importlib
import sp_cleaning_aviation as sp

importlib.reload(sp)

pd.set_option("display.max_columns", None)
```

```python
final_df = pd.read_csv('../data/joint_dataset.csv')
```

```python
# A copy of the joint dataset is created to implement the changes

aviation_df = final_df.copy()
```

```python
sp.preliminary_eda(aviation_df)
```

```python
# Drop redundant or non-informative columns

cols_to_drop = [
    "AIRLINE_x", # duplicate of AIRLINE_NAME
    "IATA_CODE", # redundant airline code
    "IATA_CODE_ORIGIN", # redundant origin airport code
    "IATA_CODE_DEST", # redundant destination airport code
    "COUNTRY", # all USA, not informative
    "COUNTRY_DEST" # all USA, not informative
]

aviation_df = aviation_df.drop(columns=cols_to_drop)
```

```python
# Rename columns for symmetry and clarity

aviation_df.rename(columns={
    "CITY": "CITY_ORIGIN",
    "STATE": "STATE_ORIGIN",
    "AIRPORT": "AIRPORT_ORIGIN"
}, inplace=True)
```

```python
# Fill missing values in numeric columns with 0

numeric_cols_zero_fill = [
    "DEPARTURE_DELAY", "ARRIVAL_DELAY", "TAXI_OUT", "TAXI_IN", 
    "AIR_TIME", "WHEELS_OFF", "WHEELS_ON",
    "AIR_SYSTEM_DELAY", "SECURITY_DELAY", "AIRLINE_DELAY",
    "LATE_AIRCRAFT_DELAY", "WEATHER_DELAY",
    "DEPARTURE_TIME", "ARRIVAL_TIME", "ELAPSED_TIME"
]

sp.zero_fill(aviation_df, numeric_cols_zero_fill)

print(aviation_df[numeric_cols_zero_fill].isna().sum())
```

```python
# Fill missing values in categorical columns with "UNKNOWN"

categorical_cols_unknown_fill = [
    "TAIL_NUMBER",
    "STATE_ORIGIN", "STATE_DEST", "CITY_ORIGIN", "CITY_DEST",
    "AIRPORT_ORIGIN", "AIRPORT_DEST"
]

sp.fill_unknown(aviation_df, categorical_cols_unknown_fill)

print(aviation_df[categorical_cols_unknown_fill].isna().sum())
```

```python
# Fill missing values for "CANCELLATION_REASON" with "None"

aviation_df["CANCELLATION_REASON"] = aviation_df["CANCELLATION_REASON"].fillna("None")
```

```python
# Delay flags

# The aim is to convert continuous delay times into categorical flags that are easier to analyze

aviation_df["DELAYED"] = np.where(aviation_df["ARRIVAL_DELAY"] > 15, 1, 0) # DELAYED = 1 if the flight arrived more than 15 minutes late, otherwise 0.
aviation_df["EXTREME_DELAY"] = np.where(aviation_df["ARRIVAL_DELAY"] > 60, 1, 0) # EXTREME_DELAY = 1 if the flight arrived more than 60 minutes late, otherwise 0.

# This will help us identify outliers or severe disruptions. 
```

```python
# Flight duration in hours

# Converting flight times from minutes to hours to make it more intuitive for analysis and easier to compare flight durations. 

aviation_df["AIR_TIME_HOURS"] = aviation_df["AIR_TIME"] / 60
aviation_df["SCHEDULED_TIME_HOURS"] = aviation_df["SCHEDULED_TIME"] / 60

aviation_df["AIR_TIME_HOURS"] = aviation_df["AIR_TIME_HOURS"].round(2)
aviation_df["SCHEDULED_TIME_HOURS"] = aviation_df["SCHEDULED_TIME_HOURS"].round(2)
```

```python
# Route feature

# Creation of a single string representing the flight route

aviation_df["ROUTE"] = aviation_df["CITY_ORIGIN"].astype(str) + "-" + aviation_df["CITY_DEST"].astype(str)
```

```python
# Date features

# Combination of year, month and day into a single datetime object to help filter.
# Extraction of the day of the week as a string to analyze patterns.
# Extraction of the month name to help detect seasonal trends.

aviation_df["DATE"] = pd.to_datetime(aviation_df[['YEAR', 'MONTH', 'DAY']])
aviation_df["WEEKDAY"] = aviation_df["DATE"].dt.day_name()
aviation_df["MONTH_NAME"] = aviation_df["DATE"].dt.month_name()
```

```python
# Clean string values

# Removing unnecessary leading or trailing whitespace from strings and converting all letters to uppercase for "TAIL_NUMBER"

aviation_df["AIRLINE_NAME"] = aviation_df["AIRLINE_NAME"].str.strip()
aviation_df["TAIL_NUMBER"] = aviation_df["TAIL_NUMBER"].str.upper()
```

```python
# Drop coordinates columns

coords_cols = ["LATITUDE", "LONGITUDE", "LATITUDE_DEST", "LONGITUDE_DEST"]
aviation_df = aviation_df.drop(columns=coords_cols)
```

```python
# Fill missing values in scheduled time columns with median

scheduled_cols = ["SCHEDULED_TIME", "SCHEDULED_TIME_HOURS"]
for col in scheduled_cols:
    median_val = aviation_df[col].median()
    aviation_df[col].fillna(median_val, inplace=True)
```

```python
# Ordenamos columnas para facilitar lectura del documento

cols_order = [
    "DATE", "YEAR", "MONTH", "DAY", "DAY_OF_WEEK", "WEEKDAY", "MONTH_NAME",
    "AIRLINE_NAME", "TAIL_NUMBER", "FLIGHT_NUMBER",
    "ORIGIN_AIRPORT", "DESTINATION_AIRPORT", 
    "SCHEDULED_DEPARTURE", "DEPARTURE_TIME", "DEPARTURE_DELAY", "TAXI_OUT", "WHEELS_OFF",
    "SCHEDULED_TIME", "SCHEDULED_TIME_HOURS", "ELAPSED_TIME", "AIR_TIME", "AIR_TIME_HOURS",
    "WHEELS_ON", "TAXI_IN", "ARRIVAL_TIME", "ARRIVAL_DELAY",
    "DELAYED", "EXTREME_DELAY",
    "CANCELLED", "DIVERTED", "CANCELLATION_REASON",
    "AIR_SYSTEM_DELAY", "SECURITY_DELAY", "AIRLINE_DELAY", "LATE_AIRCRAFT_DELAY", "WEATHER_DELAY",
    "AIRPORT_ORIGIN", "CITY_ORIGIN", "STATE_ORIGIN", "AIRPORT_DEST", "CITY_DEST", "STATE_DEST", "ROUTE"
]

aviation_df = aviation_df[cols_order]
```

```python
# Eliminamos las columnas redundantes para nuestro análisis

aviation_df.drop(columns=["DAY_OF_WEEK", "MONTH", "ORIGIN_AIRPORT", "DESTINATION_AIRPORT"], inplace=True)
```

```python
# Fixing all arrival-related and flight-execution fields for the cancelled flight. If the event did not occur, the values should be NaN (Not a Number).

aviation_df.loc[
    aviation_df["CANCELLED"] == 1,
    [
        "ARRIVAL_DELAY",
        "AIR_TIME",
        "AIR_TIME_HOURS",
        "ELAPSED_TIME",
        "WHEELS_ON",
        "ARRIVAL_TIME",
        "TAXI_IN"
    ]
] = np.nan
```

```python
aviation_df.loc[
    aviation_df["CANCELLED"] == 1,
    ["DELAYED", "EXTREME_DELAY"]
] = np.nan
```

```python
# For the diverted flights, we are going to treat them like the cancelled flight as the plane didn't arrive at the chosen destination. 

aviation_df.loc[
    aviation_df["DIVERTED"] == 1,
    [
        "ARRIVAL_DELAY",
        "ARRIVAL_TIME",
        "DELAYED",
        "EXTREME_DELAY"
    ]
] = np.nan
```

```python
aviation_df.loc[
    aviation_df["DIVERTED"] == 1,
    ["DELAYED", "EXTREME_DELAY"]
] = np.nan
```

```python
aviation_df.loc[aviation_df["CANCELLED"] == 1, ["DEPARTURE_TIME", "ARRIVAL_TIME"]] = np.nan
```
